# Impport Modules

```python
import pandas as pd
import numpy as np

import time
import sys

from ML_toolbox import data_describe as dd
```

# Load Data

```python
path = 'data/'
file = ['all_data']

result_path = '../result/'
result_file = ['brand','category', 'brand_category']


df_all = pd.read_csv(f'{path}{file[0]}.csv', encoding='utf8')

print(df_all.shape)
print(df_all.columns)
df_all.head(5)
```

```python
# describe data

dd.data_describe(df_all)
```

# Calculate Rating
## 1. Brand Rating

```python
# count each member brand pair purchase times

df_brand_rating = df_all[['member_id','brand','cnt']].groupby(['member_id','brand'])['cnt'].sum()
df_brand_rating.head(10)
```

```python
df_brand_rating = df_brand_rating.reset_index(inplace=False, level=['member_id','brand'])  
print(df_brand_rating.shape)
df_brand_rating.head(10)
```

## 2. Category Rating

```python
# count each member category pair purchase times

df_cat_rating = df_all[['member_id','category','cnt']].groupby(['member_id','category'])['cnt'].sum()
df_cat_rating.head(10)
```

```python
df_cat_rating = df_cat_rating.reset_index(inplace=False, level=['member_id','category'])  
print(df_cat_rating.shape)
df_cat_rating.head(10)
```

## 3. Category & Brand Rating

```python
# count each member category pair purchase times

df_cb_rating = df_all[['member_id','category','brand','cnt']].groupby(['member_id','category','brand'])['cnt'].sum()
df_cb_rating.head(10)
```

```python
df_cb_rating = df_cb_rating.reset_index(inplace=False, level=['member_id','category','brand'])  
print(df_cb_rating.shape)
df_cb_rating.head(10)
```

```python
df_cb_rating.category = df_cb_rating.category.astype(str)

df_cb_rating['item'] = df_cb_rating.category + df_cb_rating.brand

df_cb_rating = df_cb_rating.reindex(columns=['member_id', 'item','category', 'brand', 'cnt'])
df_cb_rating.head()
```

```python
# This table is used to record the new id's brand and category 

cat_brand_table = df_cb_rating.drop(['member_id', 'cnt'],axis=1)
cat_brand_table.drop_duplicates();
cat_brand_table.head(10)
```

```python
# This table is used to do recommand

df_cb_rating_new = df_cb_rating.drop(['category','brand'], axis=1)
df_cb_rating_new.head()
```
